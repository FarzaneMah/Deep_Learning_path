## Key concepts on Deep Neural Networks

1. "cache" records values from the forward propagation units and sends it to the backward propagation units because it is needed to compute the chain rule derivatives.

2. Deeper layers of nn are typically computing more complex features.

3. the number of layers is counted as the number of hidden layers + 1. The input and output layers are not counted as hidden layers.

4. during backpropagation you need to know which activation was used in the forward propagation to be able to compute the correct derivative.



5. To compute the function using a shallow network circuit, you will need a large network (where we measure size by the number of logic gates in the network), but (ii) To compute it using a deep network circuit, you need only an exponentially smaller network.
